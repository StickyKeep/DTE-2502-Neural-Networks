# Support Vector Machine Classifier

## ELI5 Explanation:

Imagine you have two types of fruits on a table: apples and oranges. You want to separate them using a straight stick so that all apples are on one side and all oranges are on the other. Support Vector Machine (SVM) does a similar job but in a mathematical way. It finds the "best" line (in 2D), plane (in 3D), or hyperplane (in more than three dimensions!) that separates different classes of data points.

## ELI Second-Year CompSci Student:

### Definition:

A Support Vector Machine (SVM) is a type of supervised learning algorithm used mainly for classification (but also regression, etc.). Given a dataset, it tries to find the optimal hyperplane which separates different classes in the feature space.

### How it Works:

1. **Feature Space**: The first step is representing your data in a feature space, which is usually \( n \)-dimensional, where \( n \) is the number of features for each data point.

2. **Find Support Vectors**: Support Vectors are the data points that are closest to the separating hyperplane. Essentially, they are the points you'd get wrong if you moved the line just a bit.

3. **Optimize Hyperplane**: The optimal hyperplane maximizes the margin, which is the distance between the nearest points (or Support Vectors) of different classes. Mathematically, the margin \( M \) is maximized, subject to certain constraints:

    $$
    \max_{w, b} M
    $$
    $$
    \text{subject to } y_i(w \cdot x_i + b) \geq M, \forall i
    $$

    Here, $ w $ is the normal vector to the hyperplane, $ b $ is the bias term, and $ x_i, y_i $ are the feature vectors and labels respectively.

4. **Classification**: A new data point is classified based on which side of the hyperplane it falls on. The decision function is generally:

    $$
    f(x) = w \cdot x + b
    $$

### What is it Used For:

- **Text Classification**: SVMs can be used to categorize emails, articles, or web pages.
- **Image Recognition**: SVMs are used for handwritten digit or face recognition.
- **Bioinformatics**: Identifying protein or gene classes.
- **Financial Forecasting**: For predicting market trends.

### Advantages:

- Effective in high-dimensional spaces.
- Works well when the dataset is linearly separable or close to it.
- It only uses a subset of training points (Support Vectors), so it's memory efficient.

### Limitations:

- Not suitable for large datasets due to high training time.
- Less effective when the data set has more noise i.e., target classes are overlapping.



# Vector Identities, Projections, and Their Applications

## Vector Identities

Vector identities are mathematical equations that relate vectors in various ways, often simplifying the representation of more complex equations. Here are some common vector identities:

1. **Commutative Property**: $ \vec{A} + \vec{B} = \vec{B} + \vec{A} $
2. **Distributive Property**: $ a(\vec{A} + \vec{B}) = a\vec{A} + a\vec{B} $
3. **Dot Product Distributive**: $ \vec{A} \cdot (\vec{B} + \vec{C}) = \vec{A} \cdot \vec{B} + \vec{A} \cdot \vec{C} $
4. **Cross Product Distributive**: $ \vec{A} \times (\vec{B} + \vec{C}) = \vec{A} \times \vec{B} + \vec{A} \times \vec{C} $

### Projection

Projection of one vector onto another is essentially finding the component of the first vector that lies in the direction of the second vector.

- **Projection Formula**: 
    $$
    \text{proj}_{\vec{B}} \vec{A} = \left( \frac{\vec{A} \cdot \vec{B}}{\vec{B} \cdot \vec{B}} \right) \vec{B}
    $$

This formula tells us how much of $ \vec{A} $ lies in the direction of $ \vec{B} $.

## Why They Are Handy

1. **Simplification**: Vector identities can simplify complex equations, making it easier to solve them.
2. **Physical Understanding**: Projections are useful for breaking down forces in physics into components that are easier to analyze.
3. **Data Science**: They are widely used in machine learning algorithms like PCA, and SVM to simplify the feature space.
4. **Computer Graphics**: In rendering and image processing, projections are used to map 3D space onto a 2D screen.

## Example

Consider vectors $ \vec{A} = [1, 2] $ and $ \vec{B} = [2, 1] $.

- To find the projection of $ \vec{A} $ onto $ \vec{B} $:
    $$
    \text{proj}_{\vec{B}} \vec{A} = \left( \frac{[1, 2] \cdot [2, 1]}{[2, 1] \cdot [2, 1]} \right) [2, 1]
    $$
    $$
    = \left( \frac{1*2 + 2*1}{2*2 + 1*1} \right) [2, 1]
    $$
    $$
    = \left( \frac{4}{5} \right) [2, 1] = \left[ \frac{8}{5}, \frac{4}{5} \right]
    $$

Here, $ \left[ \frac{8}{5}, \frac{4}{5} \right] $ is the component of $ \vec{A} $ that lies in the direction of $ \vec{B} $.


## Concrete Example

Let's consider a real-world scenario to explain vector projection. Imagine you are pulling a suitcase at an angle, and you want to find out how much force you are applying in the forward direction.

Let:
- $ \vec{A} = [3, 4] $ represent the force you are applying at an angle.
- $ \vec{B} = [1, 0] $ represent the forward direction.

### Steps to Find Projection

1. **Find $ \vec{A} \cdot \vec{B} $**
    $$
    \vec{A} \cdot \vec{B} = 3 \times 1 + 4 \times 0 = 3
    $$

2. **Find $ \vec{B} \cdot \vec{B} $**
    $$
    \vec{B} \cdot \vec{B} = 1 \times 1 + 0 \times 0 = 1
    $$

3. **Calculate the projection of $ \vec{A} $ onto $ \vec{B} $**
    $$
    \text{proj}_{\vec{B}} \vec{A} = \left( \frac{3}{1} \right) [1, 0] = [3, 0]
    $$

The vector $ [3, 0] $ tells us that we are effectively applying a force of 3 units in the forward direction.

In this example, the projection helps us understand how much of the force you are exerting is actually useful for moving the suitcase forward.

# Recap of Vector Identities in the Plane

## Addition and Subtraction

### Vector Addition
For vectors $ \vec{A} = [a_1, a_2] $ and $ \vec{B} = [b_1, b_2] $, vector addition is defined as:

$$
\vec{A} + \vec{B} = [a_1 + b_1, a_2 + b_2]
$$

### Vector Subtraction
Vector subtraction is similarly defined as:

$$
\vec{A} - \vec{B} = [a_1 - b_1, a_2 - b_2]
$$

## Scalar Multiplication

Given a vector $ \vec{A} = [a_1, a_2] $ and a scalar $ c $, scalar multiplication is:

$$
c \vec{A} = [c \cdot a_1, c \cdot a_2]
$$

## Dot Product

The dot product of two vectors $ \vec{A} = [a_1, a_2] $ and $ \vec{B} = [b_1, b_2] $ in the plane is:

$$
\vec{A} \cdot \vec{B} = a_1 \cdot b_1 + a_2 \cdot b_2
$$

## Norm (Magnitude)

The norm or magnitude of a vector $ \vec{A} = [a_1, a_2] $ is defined as:

$$
|| \vec{A} || = \sqrt{a_1^2 + a_2^2}
$$

## Unit Vector

A unit vector $ \hat{A} $ in the direction of $ \vec{A} $ is given by:

$$
\hat{A} = \frac{\vec{A}}{|| \vec{A} ||}
$$

## Orthogonal Vectors

Two vectors $ \vec{A} $ and $ \vec{B} $ are orthogonal if their dot product is zero:

$$
\vec{A} \cdot \vec{B} = 0
$$

## Miscellaneous Identities

### Distributive Law for Dot Product
$ \vec{A} \cdot (\vec{B} + \vec{C}) = \vec{A} \cdot \vec{B} + \vec{A} \cdot \vec{C} $

### Scalar Multiplication and Dot Product
$ (c \vec{A}) \cdot \vec{B} = c (\vec{A} \cdot \vec{B}) $

These are some of the fundamental vector identities and operations in the plane, useful for various applications in physics, engineering, computer science, and other fields.


# Equation of a Hyperplane and its Relations

## What is a Hyperplane?

A hyperplane is a flat affine subspace of one dimension less than its ambient space. In three dimensions, a hyperplane is a flat two-dimensional "plane" that extends infinitely. In $n$ dimensions, a hyperplane is an $(n-1)$-dimensional flat affine space.

## Equation of a Hyperplane

The equation of a hyperplane in $n$-dimensional space can be represented as:

$$
\vec{W}^T \vec{X} + b = 0
$$

Here:
- $\vec{W}$ is a weight vector in $n$ dimensions.
- $\vec{X}$ is the input vector in $n$ dimensions.
- $b$ is a bias term.
- $\vec{W}^T$ indicates the transpose of $\vec{W}$.
  
## How It Relates to $y = ax + b$

In two dimensions, the equation of a line is given by $y = ax + b$. This can be rewritten in vector form as:

$$
\begin{pmatrix} -a \\ 1 \end{pmatrix} \cdot \begin{pmatrix} x \\ y \end{pmatrix} = -b
$$

Or:

$$
\begin{pmatrix} a \\ -1 \end{pmatrix} \cdot \begin{pmatrix} x \\ y \end{pmatrix} = b
$$

In this specific case, $\vec{W} = \begin{pmatrix} a \\ -1 \end{pmatrix}$, $\vec{X} = \begin{pmatrix} x \\ y \end{pmatrix}$, and $b$ remains the same, satisfying the equation $\vec{W}^T \vec{X} + b = 0$.

## How It Relates to $W^T \times X = 0$

The equation $W^T \times X = 0$ is a special case of a hyperplane passing through the origin. When a hyperplane passes through the origin, the bias term $b$ becomes zero, simplifying the general equation to:

$$
\vec{W}^T \vec{X} = 0
$$

## Conclusion

The equation of a hyperplane is a natural extension of the equation of a line in two dimensions. It generalizes the concept by introducing a weight vector and a bias term, which also allows it to represent planes in higher dimensions. The equation $W^T \times X = 0$ and the 2D line equation $y = ax + b$ can be considered as specific cases of the general hyperplane equation.


# Support Vector Machines (SVM) and Hyperplanes: A Comprehensive Guide

## For Beginners: What is a Support Vector Machine (SVM)?

Imagine you have a piece of graph paper filled with red and blue dots. Your task is to draw a straight line that best separates the dots into their respective colors. That's pretty much what a Support Vector Machine (SVM) does—except it does it computationally, often with data that is much more complex than just colored dots.

## For CompSci Students: Algorithmic Overview

SVM is an algorithmic approach for supervised learning, which is a type of machine learning. It's mainly used for classification tasks but can also perform regression (known as SVR, Support Vector Regression). The core idea is to find the "optimal" hyperplane that distinctly classifies the data points in multi-dimensional space.

## Hyperplanes: What Are They?

### For Beginners:

Think of a hyperplane as a fancy name for something very simple. In a 2D world, a hyperplane is just a line. If you go up to a 3D world, it becomes a flat sheet. 

### For CompSci Students:

In an $n$-dimensional space, a hyperplane is an $(n-1)$-dimensional flat affine subspace. For example, in a 3D space, a hyperplane is a 2D plane. Mathematically, the general equation of a hyperplane is:

$$
\vec{W}^T \vec{X} + b = 0
$$

Here, $\vec{W}$ is the weight vector and $b$ is the bias term. $\vec{X}$ is the input data point, and $\vec{W}^T$ is the transpose of $\vec{W}$.

## Equations: How to Represent a Line?

### For Beginners:

Every line can be described with an equation that looks like $y = mx + c$. The $m$ tells us how slanted the line is, and the $c$ tells us where it crosses the vertical axis.

### For CompSci Students:

In vector form, the equation of a line in 2D can be extended to represent hyperplanes in higher dimensions. For a line in 2D, the equation can be written in vector notation as:

$$
\vec{W}^T \vec{X} = b
$$

Here, $\vec{W} = [m, -1]$ and $\vec{X} = [x, y]$, making it a specific case of the general hyperplane equation.

## The Algorithm: Finding the Best Line

### For Beginners:

SVM tries to find the best line by making sure it's as far away from the closest red and blue dots as possible. This ensures better decision-making for new data points.

### For CompSci Students:

SVM uses a method called convex optimization to find the values of $\vec{W}$ and $b$ that maximize the margin between the two classes. The margin is the distance between the closest points (known as support vectors) to the hyperplane from each class.

## Summary

SVM is essentially a high-powered version of drawing a line to separate red and blue dots. It's an essential tool in machine learning and is rooted in optimizing complex equations. Whether you are new to the subject or a first-year computer science student, understanding the basics of SVM and hyperplanes can offer a solid foundation for more advanced machine learning techniques.


# An Easy-ish Example Implementations in Python

Generated by GPT4

How Does It Work?
Dataset Creation: We create a small dataset X with 5 data points and their corresponding labels y. The points with label 0 are one class, and those with label 1 are another.

* SVM Model: We use scikit-learn's SVC class to create an SVM model with a linear kernel.

* Model Fitting: We fit the model using our dataset.

* Visualization: The script visualizes the dataset points, the best separating line, and the margins (dotted lines).

* Calculating the Margin

After running the code, you'll see a plot where the dotted lines indicate the margins, and the solid line is the "best" line that separates the two classes. The large points on the plot are the "support vectors," which are the points that the algorithm uses to determine the best separating line. The calculated maximum margin will also be printed.

This should provide a straightforward example of how SVM works to find the best separating line in a 2D space.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create extended dataset
X = np.array([[1, 1],
              [1, 3],
              [2, 2],
              [2.5, 1.5],
              [3, 1],
              [3.5, 2.5],
              [2, 0],
              [3, 0],
              [3.5, 1]])

y = np.array([0, 0, 0, 0, 1, 1, 1, 1, 1])

# Create an SVM model
clf = svm.SVC(kernel='linear', C=1000)

# Fit the model
clf.fit(X, y)

# Plotting the points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', label='Data Points')

# Plot the separating hyperplane
ax = plt.gca()
xlim = ax.get_xlim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 50)
yy = np.linspace(xlim[0], xlim[1], 50)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# Plot the decision boundary and margins
contour = ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.clabel(contour, inline=True, fontsize=10)
plt.title('SVM with Hyperplane and Margins')
plt.xlabel('X')
plt.ylabel('Y')

# Plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k', label='Support Vectors')

plt.legend()
plt.show()

# Calculating the margin
margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
print(f"The maximum margin is: {margin}")
```

# Iiris Dataset Implementation Example

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]  # We only use petal length and petal width for simplicity
y = iris.target

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

# Standardize features
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

# Initialize SVM classifier
svm = SVC(kernel='linear', C=1.0, random_state=1)

# Fit model
svm.fit(X_train_std, y_train)

# Predict
y_pred = svm.predict(X_test_std)

from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=colors[idx], marker=markers[idx], label=cl, edgecolor='black')
    
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], edgecolor='black', alpha=1.0, linewidth=1, marker='o', s=100, label='test set')


# Combine standardized training and test sets
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

# Plotting the decision regions
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```
